```python
csv1["employment_type"] = (
    csv1["employment_type"]
    .str.lower()
    .str.replace("_", " ")
    .str.replace("-", " ")
)

```

```python
csv1["employment_type"] = csv1["employment_type"].replace({
    "ft": "full-time",
    "fulltime": "full-time",
    "full-time": "full-time",
    "FULL_TIME": "full-time",
    "FT": "full-time",

    "pt": "part-time",
    "Part Time": "part-time",
    "Part-time": "part-time",
    "PART_TIME": "part-time",
    "PT": "part-time",

    "se": "self-employed",
    "Self_emp": "self-employed",
    "SELF_EMPLOYED": "self-employed",
    "Self Emp": "self-employed",
    "self employed": "self-employed",

    "ct": "contract",
    "CONTRACT": "contract",
    "Contract": "contract",
    "contractor": "contract"
})
```

```python
csv1['employment_length'] = csv1['employment_length'].fillna(0)
```

```python
csv1["annual_income"] = (csv1["annual_income"].str.replace("$", "", regex=False) .str.replace(",", "", regex=False).astype(float))
```

```python
csv1["annual_income"] = csv1["annual_income"].astype(int)
```
